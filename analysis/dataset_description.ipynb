```python
import pandas as pd
import networkx as nx
from networkx.algorithms.approximation.clique import maximum_independent_set

# Load data


# Create the graph
G = nx.Graph()

# Add nodes with their attributes
for index, row in nodes_df.iterrows():
    G.add_node(row['id'], pos_x=row['pos_x'], pos_y=row['pos_y'], pos_z=row['pos_z'], degree=row['degree'], isAtSampleBorder=row['isAtSampleBorder'])

# Add edges with their attributes
for index, row in edges_df.iterrows():
    G.add_edge(row['node1id'], row['node2id'], length=row['length'], distance=row['distance'], curveness=row['curveness'],
               volume=row['volume'], avgCrossSection=row['avgCrossSection'], minRadiusAvg=row['minRadiusAvg'],
               minRadiusStd=row['minRadiusStd'], avgRadiusAvg=row['avgRadiusAvg'], avgRadiusStd=row['avgRadiusStd'],
               maxRadiusAvg=row['maxRadiusAvg'], maxRadiusStd=row['maxRadiusStd'], roundnessAvg=row['roundnessAvg'],
               roundnessStd=row['roundnessStd'], node1_degree=row['node1_degree'], node2_degree=row['node2_degree'],
               num_voxels=row['num_voxels'], hasNodeAtSampleBorder=row['hasNodeAtSampleBorder'])

# Calculate statistics
num_vertices = G.number_of_nodes()
num_edges = G.number_of_edges()
degrees = [d for n, d in G.degree()]
mean_degree = sum(degrees) / num_vertices
min_degree = min(degrees)
max_degree = max(degrees)

#independence_num = len(maximum_independent_set(G))
#two_independence_num = len(maximum_independent_set(nx.power(G, 2)))

# Prepare the LaTeX table
latex_table = r"""
\begin{table}[h!]
\centering
\begin{tabular}{|l|r|}
\hline
\textbf{Statistic} & \textbf{Value} \\
\hline
Number of Vertices & """ + f"{num_vertices}" + r""" \\
\hline
Number of Edges & """ + f"{num_edges}" + r""" \\
\hline
Mean Node Degree & """ + f"{mean_degree:.2f}" + r""" \\
\hline
Min Node Degree & """ + f"{min_degree}" + r""" \\
\hline
Max Node Degree & """ + f"{max_degree}" + r""" \\
\hline
\end{tabular}
\caption{Graph Statistics}
\label{tab:graph_stats}
\end{table}
"""

print(latex_table)
```
